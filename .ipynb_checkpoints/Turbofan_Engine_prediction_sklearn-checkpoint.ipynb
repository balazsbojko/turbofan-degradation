#### 1.3. Separating the DataFrame with respect to single engines

##### 1.3.1. Functions for choosing an engine and adding the RUL column

```python
# Creating a Pandas DataFrame from an engine
def uniqueEnginePD(pdf, serial_no):
    unique_engine_PD = pdf[pdf['Engine number'] == serial_no]
    return unique_engine_PD
```

```python
# Appending the RUL column to a Pandas DataFrame
def appendRulPD(unique_engine):
    #cycles = int(unique_engine['Cycles'].max())
    # Use the length of the DataFrame instead, some cycles may be deleted later, due to smoothing / featuring
    cycles = len(unique_engine)
    unique_engine = unique_engine.assign(RUL = list(np.arange(cycles-1, -1, -1)))
    unique_engine_RUL_PD = unique_engine.reset_index()
    del unique_engine_RUL_PD['index']
    return unique_engine_RUL_PD
```

### 2. Exploring and transforming the dataset (plotting and statistical info)

#### 2.1. Creating the plotter function

```python
# Function for plotting the 21 sensor traces. Input: Pandas DataFrame
def plotData(df):
    # for convenience, identify the sensor and operational setting columns
    sensor_columns = [col for col in df.columns if col.startswith("Sensor")]
    setting_columns = [col for col in df.columns if col.startswith("Setting")]
    
    # Plot values in time
    fig, axes = plt.subplots(7, 3, figsize = (10,10), sharex=True)
    for index, ax in enumerate(axes.ravel()):
        # plot the channel corresponding to this position
        sensor_col = sensor_columns[index]
        
        df.plot(x='Cycles', y=sensor_col, ax=ax, color='black')
        # label formatting
        if index % 3 == 0:
            ax.set_ylabel("Sensor Value", size=10);
        else:
            ax.set_ylabel("");
        ax.set_xlabel("Time (Cycles)");
        ax.set_title(sensor_col.title(), size=14);
        ax.legend_.remove();
    # plot formatting
    fig.suptitle("Sensor Traces : Unit "+ str(df['Engine number'][0]), size=20, y=1.025)
    fig.tight_layout()
    
    # Show plot
    plt.show()
```

#### 2.3. Create a new feature set and training set for the rolling average data of a single engine

```python
# This function smoothes the sensor data with the rolling average with the specified cycle window
def generateRollingDF(df, smoothing_window, indices_rolling):
    
    # Copy the input DataFrame
    df_rolling = df.copy()
    
    # Compute rolling average
    df_rolling[indices_rolling] = df[indices_rolling].rolling(smoothing_window).mean()
    
    # Delete all rows from the table containing NaNs (assuming previous data cleaning and the effect of
    # rolling avg. resulting in NaNs)
    df_rolling.dropna(axis = 'index', how = 'any', inplace = True)
    
    df_rolling = df_rolling.reset_index()
    del df_rolling['index']
    
    return df_rolling
```

```python
# Classify RUL and create a new column in the DataFrame. The warning variable 
def classifyRUL(df, warning):
    string = 'Fail label ' + str(warning) + ' s'
    df_classified = df.copy()
    df_classified[string] = (df['RUL'] <= warning).astype(int)
    return df_classified
```

```python
# Choose value of windowing for the sensor value smoothing (smoothing = 10)
engine45_pd_rf = generateRollingDF(engine45_pd, 10, indices_rolling)

# 20 s before failure is the beginning of the warning (warning = 20 cycles before failure)
df45_rf_class = classifyRUL(engine45_pd_rf, 20)
```

##### 3.2.2. Function for predicting results for a particular engine

```python
# Input: test set DataFrame. Returns a Pandas DataFrame with the predicted result.
def predictEngine_pd(test_df, engine_no, ML_model):
    # Choose the engine, convert to Pandas DataFrame
    test_engine_pd = uniqueEnginePD(test_df, engine_no).copy()
    
    # Compute the predictions on the test set
    pred_engine_pd = ML_model.predict(test_engine_pd.iloc[:, 2:])
    
    test_engine_pd['predict'] = pred_engine_pd
    
    return test_engine_pd
```

##### 3.3.6. Add real RUL to the predicted dataframe

```python
def predictEngineWithRUL(test_df, RUL_test_df, engine_no, ML_model, warning):
    # Predict for an engine
    test_engine_predicted = predictEngine_pd(test_df, engine_no, ML_model)
    
    # Get RUL for the particular engine
    rul = RUL_test_df['RUL_test'][RUL_test_df['Engine number'] == engine_no].values[0]
    
    test_engine_predicted['RUL_remaining'] = 0
    
    test_engine_predicted.reset_index(inplace=True, drop=True)
    
    # How many cycles must be predicted as 1 at the tail of the data?
    remaining = int(warning-rul)
    start = len(test_engine_predicted)-remaining
    
    # If remaining is greater than zero, add '1' values to the RUL_remaining
    if remaining > 0:
        test_engine_predicted.loc[start:,'RUL_remaining'] = 1
    
    return test_engine_predicted
```

#### 4.2. Windowed feature creation

```python
from timeit import default_timer as timer
```

```python
# Performance optimized feature creation. Works only if there are exactly 2 columns before features!
def featureWindowing(df, features, window):
    # Make a copy to keep input dataframe unmodified
    df_new = df.copy()
    
    start = timer()
    # Create columns according to the time window
    for i in range(1, window):
        for j in features:
            sensor_name = j + ' t-' + str(i)
            df_new[sensor_name] = np.nan
    
    # Fill new features with previous 'window' cycle data
    for k in range(window-1, len(df)):  # k will be the row index (starting from 0) from the beginning
        for l in range(1, window):
            df_new.loc[k, (len(features)*l + 2):(len(features)*(l+1) + 2)] = np.asarray(df.iloc[k-l, 2:(len(features) + 2)])                
    end = timer()
    print('Feature creation time = ', end-start)
    
    # Delete the rows containing NaN values - option for later: predict on rows containing NaNs    
    df_new.dropna(axis = 'index', how = 'any', inplace = True)
    
    df_new = df_new.reset_index()
    del df_new['index']
    
    return df_new
```

```python
# Create one dataframe with all engines, prepared for ML training

# Measure time
start_assembly = timer()

# Get the number of engines in training set
no_of_engines = int(train_FD001['Engine number'].max())

# Smooth the training set for the first engine in the dataset
smooth_training_set = generateRollingDF(uniqueEnginePD(train_FD001, 1), smoothing, indices_rolling)

# Do the windowing computation
windowed_training_set = featureWindowing(smooth_training_set, features, time_windowing)

# Append RUL
windowed_training_set_RUL = appendRulPD(windowed_training_set)

for i in range(2, no_of_engines + 1):
    # Merge
    smooth_training_set_temp = generateRollingDF(uniqueEnginePD(train_FD001, i), smoothing, indices_rolling)
    
    windowed_training_set_temp = featureWindowing(smooth_training_set_temp, features, time_windowing)
    
    windowed_training_set_RUL_temp = appendRulPD(windowed_training_set_temp)
    
    windowed_training_set_RUL = windowed_training_set_RUL.append(windowed_training_set_RUL_temp)

# Classify RUL according to the specified warning limit
windowed_training_set_cl = classifyRUL(windowed_training_set_RUL, warning)

end_assembly = timer()
print('Total feature creation time = ', (end_assembly-start_assembly))
```
